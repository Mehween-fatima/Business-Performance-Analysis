```python
import pandas as pd

summary = {}

# 1. Overall performance
summary["Total Sales"] = df['sales'].sum()
summary["Total Profit"] = df['profit'].sum()
summary["Profit Margin (%)"] = (df['profit'].sum() / df['sales'].sum()) * 100

# 2. Loss analysis
summary["Loss-Making Orders"] = (df['profit'] < 0).sum()
summary["Loss Percentage"] = ((df['profit'] < 0).mean()) * 100

# 3. Segment performance
segment_perf = df.groupby('segment')[['sales','profit']].sum()
summary["Worst Segment (Profit)"] = segment_perf['profit'].idxmin()
summary["Best Segment (Profit)"] = segment_perf['profit'].idxmax()

# 4. Product / Sub-category
subcat_perf = df.groupby('sub-category')['profit'].sum()
summary["Worst Sub-Category"] = subcat_perf.idxmin()
summary["Best Sub-Category"] = subcat_perf.idxmax()

# 5. Region / State (auto-detect)
if 'state' in df.columns:
    state_perf = df.groupby('state')['profit'].sum()
    summary["Worst State"] = state_perf.idxmin()
    summary["Best State"] = state_perf.idxmax()

# 6. Discount impact
summary["Average Discount"] = df['discount'].mean()
summary["High Discount Loss Orders"] = (
    df[df['discount'] > df['discount'].median()]['profit'] < 0
).mean() * 100

# 7. Pareto Analysis
product_profit = df.groupby('product_name')['profit'].sum().sort_values(ascending=False)
cum_profit_pct = product_profit.cumsum() / product_profit.sum()
summary["Products Driving 80% Profit"] = (cum_profit_pct <= 0.8).sum()

# 8. Profit simulation (5% discount cut)
summary["Profit Increase if Discount Cut 5% (%)"] = (
    ((df['profit'] + (df['sales'] * 0.05)).sum() - df['profit'].sum())
    / df['profit'].sum()
) * 100


# 🧠 AUTO INSIGHT PRINTER
print("\n📊 AUTOMATIC BUSINESS SUMMARY")
print("=" * 50)
for k, v in summary.items():
    if isinstance(v, float):
        print(f"{k}: {v:,.2f}")
    else:
        print(f"{k}: {v}")
print("=" * 50)
```

```python
pd.read_sql("""
SELECT 
    State,
    SUM(Sales) AS total_sales,
    SUM(Profit) AS total_profit
FROM superstore_sales
GROUP BY State
ORDER BY total_profit ASC
""", conn).head(3)
```

The analysis reveals that texas,ohio,pennsylvania by total profit are generating significant losses despite recording sales activity.

```python
pd.read_sql("""
SELECT 
    "Order ID",
    "Sales",
    "Profit",
    CASE
        WHEN "Profit" < 0 THEN 'Loss'
        WHEN "Profit" BETWEEN 0 AND 100 THEN 'Low Profit'
        ELSE 'High Profit'
    END AS profit_status
FROM superstore_sales
""", conn).head(5)
```

its identified that these are the loss making,and high profit orders

```python
pd.read_sql("""
SELECT 
    "sub-category",
    "Category",
    SUM(Sales) AS total_sales,
    RANK() OVER (
        PARTITION BY "Category" 
        ORDER BY SUM(Sales) DESC
    ) AS sales_rank
FROM superstore_sales
GROUP BY "sub-category", "Category"
""", conn).head(5)
```

Furniture is the top performing category and chairs are the top performing sub category

```python
pd.read_sql("""
SELECT 
    "category",
    "Sales"
FROM superstore_sales
WHERE Sales > (
    SELECT AVG(Sales)
    FROM superstore_sales
)
""", conn).head(5)
```

it is identified that furniture sells better than average and is the highest profitable category

```python
pd.read_sql("""
SELECT 
    "Order Date",
    "Sales",
    SUM(Sales) OVER (
        ORDER BY "Order Date"
    ) AS cumulative_sales
FROM superstore_sales
""", conn).head(5)
```

Cumulative sales show a steady upward trend over time, with revenue growth driven by multiple transactions occurring on the same dates rather than isolated large orders.

```python
overview = pd.read_sql("""
SELECT
    COUNT(DISTINCT "Customer Name") AS total_customers,
    COUNT(DISTINCT State) AS total_states,
    COUNT(DISTINCT Category) AS total_categories,
    COUNT(DISTINCT "Order ID") AS total_orders,
    MIN("Order Date") AS start_date,
    MAX("Order Date") AS end_date,
    ROUND(SUM(Sales), 2) AS total_sales,
    ROUND(SUM(Profit), 2) AS total_profit,
    ROUND(AVG(Profit), 2) AS avg_profit
FROM superstore_sales
""", conn)

overview
```

```python
loss_stats = pd.read_sql("""
SELECT
    SUM(CASE WHEN Profit < 0 THEN 1 ELSE 0 END) AS loss_orders,
    ROUND(AVG(Profit), 2) AS avg_profit
FROM superstore_sales
""", conn)

top_region = pd.read_sql("""
SELECT Region, SUM(Sales) AS sales
FROM superstore_sales
GROUP BY Region
ORDER BY sales DESC
LIMIT 1
""", conn)
```
